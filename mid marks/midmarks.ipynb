```python
import pandas as pd
df = pd.read_csv("MIDMARKS.csv")
df = df.dropna(how='all')
subjects = ['DV', 'M-II', 'PP', 'BEEE', 'FL', 'FIMS']
df[subjects] = df[subjects].apply(pd.to_numeric, errors='coerce').fillna(0)
if 'SECTION' in df.columns:
   
    df['SECTION'] = df['SECTION'].str.strip().str.lower()

    # Function to calculate pass/fail stats for each subject in a section
    def calculate_probabilities(group):
        stats = {'Total Students': len(group)}
        for subject in subjects:
            pass_count = (group[subject] >= 10).sum()
            fail_count = (group[subject] < 10).sum()
            stats[f'{subject} Pass Probability'] = pass_count / len(group)
            stats[f'{subject} Fail Probability'] = fail_count / len(group)
        return pd.Series(stats)

    # Group by Section and apply the function
    section_subject_probabilities = df.groupby('SECTION').apply(calculate_probabilities)

    # Display results
    print(section_subject_probabilities)
else:
    print("No 'SECTION' column found in the dataset.")
```

```python
import pandas as pd
k = pd.read_csv("MIDMARKS.csv")
for col in k.columns[2:]:
    k[col] = pd.to_numeric(k[col], errors='coerce')
passing_marks = 10
subject_section_pass = k.groupby('SECTION').apply(lambda df: (df.iloc[:, 2:] >= passing_marks).sum())
subject_section_fail = k.groupby('SECTION').apply(lambda df: (df.iloc[:, 2:] < passing_marks).sum())
total_students_per_section = k.groupby('SECTION').size()
prob_pass = subject_section_pass.div(total_students_per_section, axis=0)
prob_fail = subject_section_fail.div(total_students_per_section, axis=0)
subject_section_pass['Total Pass'] = subject_section_pass.sum(axis=1)
subject_section_fail['Total Fail'] = subject_section_fail.sum(axis=1)
prob_pass['Average Pass Probability'] = prob_pass.mean(axis=1)
prob_fail['Average Fail Probability'] = prob_fail.mean(axis=1)
subject_section_pass.loc['Total'] = subject_section_pass.sum()
subject_section_fail.loc['Total'] = subject_section_fail.sum()
prob_pass.loc['Total'] = prob_pass.mean()
prob_fail.loc['Total'] = prob_fail.mean()
print("Number of students passing each subject in each section with Totals:")
print(subject_section_pass)
print("\nNumber of students failing each subject in each section with Totals:")
print(subject_section_fail)
print("\nProbability of students passing each subject in each section with Averages (0 to 1):")
print(prob_pass)
print("\nProbability of students failing each subject in each section with Averages (0 to 1):")
print(prob_fail)
```
