# Bar plot: Total production vs. dispatch for each company

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
df = pd.read_csv("CoalProduction_Dispatch_2021-22_1.csv")
```

```python
plt.figure(figsize=(12, 6))
sns.barplot(x="Coal Company", y="Cumulative", hue="Coal Production/Dispatch", data=df)
plt.xticks(rotation=45)
plt.title("Total Production vs. Dispatch for Each Coal Company")
plt.xlabel("Coal Company")
plt.ylabel("Total Coal (Million Tonnes)")
plt.legend(title="Category")
plt.show()
```

#  Line plot: Monthly production trend of a selected company

```python
ecl_df = df[(df["Coal Company"] == "ECL") & (df["Coal Production/Dispatch"] == "PRODUCTION")]

plt.figure(figsize=(10, 5))
months = df.columns[3:-1]  # Extract month columns
plt.plot(months, ecl_df.iloc[0, 3:-1], marker="o", linestyle="-", label="ECL Production")
plt.xticks(rotation=45)
plt.title("ECL Monthly Production Trend")
plt.xlabel("Month")
plt.ylabel("Coal Production (Million Tonnes)")
plt.legend()
plt.grid()
plt.show()
```

#  Heatmap: Monthly coal production across companies

```python
prod_df = df[df["Coal Production/Dispatch"] == "PRODUCTION"].set_index("Coal Company")[df.columns[3:-1]]

plt.figure(figsize=(12, 6))
sns.heatmap(prod_df, cmap="Blues", annot=True, fmt=".1f")
plt.title("Monthly Coal Production Across Companies")
plt.xlabel("Month")
plt.ylabel("Coal Company")
plt.show()
```

# Pie Chart: Share of each company in total production

```python
prod_df = df[df["Coal Production/Dispatch"] == "PRODUCTION"]
plt.figure(figsize=(8, 8))
plt.pie(prod_df["Cumulative"], labels=prod_df["Coal Company"], autopct="%1.1f%%", startangle=140)
plt.title("Share of Each Company in Total Production")
plt.show()
```

# Box Plot: Distribution of monthly production values

```python
plt.figure(figsize=(10, 5))
sns.boxplot(data=prod_df[df.columns[3:-1]])
plt.xticks(rotation=45)
plt.title("Distribution of Monthly Production")
plt.xlabel("Month")
plt.ylabel("Coal Production (Million Tonnes)")
plt.show()
```

#  Scatter Plot: Relationship between Target and Cumulative production

```python
plt.figure(figsize=(8, 6))
sns.scatterplot(x=df["Target"], y=df["Cumulative"], hue=df["Coal Company"], s=100)
plt.title("Target vs. Cumulative Production")
plt.xlabel("Target (Million Tonnes)")
plt.ylabel("Cumulative Production (Million Tonnes)")
plt.grid()
plt.show()
```

# Line Plot: Total monthly production trend

```python
monthly_production = df[df["Coal Production/Dispatch"] == "PRODUCTION"].sum(numeric_only=True)

plt.figure(figsize=(10, 5))
plt.plot(monthly_production.index, monthly_production.values, marker="o", linestyle="-", color="b")
plt.xticks(rotation=45)
plt.title("Total Monthly Production Trend")
plt.xlabel("Month")
plt.ylabel("Total Coal Production (Million Tonnes)")
plt.grid()
plt.show()
```

#  Stacked Bar Chart: Monthly production of each company

```python
import matplotlib.pyplot as plt
import pandas as pd
df = pd.read_csv("CoalProduction_Dispatch_2021-22_1.csv")
prod_df = df[df["Coal Production/Dispatch"] == "PRODUCTION"]
prod_df_numeric = prod_df[df.columns[3:-1]]  # Selecting only month columns
plt.figure(figsize=(12, 6))
prod_df_numeric.T.plot(kind="bar", stacked=True, figsize=(12, 6), colormap="coolwarm")
plt.xticks(rotation=45)
plt.title("Stacked Bar Chart of Monthly Production by Company")
plt.xlabel("Month")
plt.ylabel("Coal Production (Million Tonnes)")
plt.legend(title="Coal Company", bbox_to_anchor=(1.05, 1), loc="upper left")
plt.show()
```

# Violin Plot: Monthly Production Distribution

```python
plt.figure(figsize=(10, 5))
sns.violinplot(data=prod_df[df.columns[3:-1]])
plt.xticks(rotation=45)
plt.title("Violin Plot of Monthly Production")
plt.xlabel("Month")
plt.ylabel("Coal Production (Million Tonnes)")
plt.show()
```
